```python
from pathlib import Path
import sys

try:
    from google.colab import drive  # type: ignore[import-not-found]

    drive.mount("/content/drive")
except ModuleNotFoundError:
    pass

bootstrap_candidates = [
    Path("/content/drive/MyDrive/basketball_hackathon/course"),
    Path.cwd().resolve(),
    *Path.cwd().resolve().parents,
]
for candidate in bootstrap_candidates:
    if (candidate / "src" / "course_setup.py").exists():
        if str(candidate) not in sys.path:
            sys.path.insert(0, str(candidate))
        break
else:
    raise FileNotFoundError(
        "找不到 src/course_setup.py。請先執行 init_colab.ipynb，確認課程已同步到 /content/drive/MyDrive/basketball_hackathon/course。"
    )

from src.course_setup import DEFAULT_COURSE_ROOT, bootstrap_course_notebook  # noqa: E402

COURSE_ROOT = bootstrap_course_notebook(DEFAULT_COURSE_ROOT, mount_drive=False)
```

```python
import shutil
import zipfile
from pathlib import Path

RESULTS = COURSE_ROOT / "assets" / "results"
SHOWCASE = RESULTS / "showcase"
SHOWCASE.mkdir(parents=True, exist_ok=True)

patterns = [
    "d5_*.png",
    "d5_*.json",
    "d4_02_pose_angle_plot.png",
    "d4_03_ball_path.png",
]

copied = []
for pattern in patterns:
    for p in RESULTS.glob(pattern):
        target = SHOWCASE / p.name
        shutil.copy(p, target)
        copied.append(target)

print("copied files:")
for p in copied:
    print("-", p.name)
```

```python
zip_path = RESULTS / "student_showcase.zip"
if zip_path.exists():
    zip_path.unlink()

with zipfile.ZipFile(zip_path, "w", compression=zipfile.ZIP_DEFLATED) as zf:
    for p in SHOWCASE.glob("*"):
        zf.write(p, arcname=p.name)

print("showcase zip:", zip_path)
```
